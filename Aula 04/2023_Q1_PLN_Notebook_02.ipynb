```python
import csv

# nome da conta do Twitter que voc√™ deseja buscar os seguidores
nome_conta = "adalves17"

# obtenha os IDs dos seguidores da conta
seguidores_ids = tweepy.Cursor(api.followers_ids, screen_name=nome_conta).items()

# abra um arquivo CSV para escrever as informa√ß√µes dos seguidores
with open("seguidores.csv", mode="w", encoding="utf-8", newline="") as arquivo_csv:
    # crie um escritor CSV para escrever as informa√ß√µes dos seguidores no arquivo
    escritor_csv = csv.writer(arquivo_csv)
    # escreva o cabe√ßalho do arquivo CSV
    escritor_csv.writerow(["ID", "Nome", "Nome de usu√°rio", "Localiza√ß√£o", "Descri√ß√£o", "N√∫mero de seguidores", "N√∫mero de amigos"])

    # processe cada seguidor retornado pela busca
    for seguidor_id in seguidores_ids:
        # obtenha as informa√ß√µes do seguidor a partir do ID
        seguidor = api.get_user(seguidor_id)

        # extraia as informa√ß√µes relevantes do seguidor
        id_seguidor = seguidor.id_str
        nome_seguidor = seguidor.name
        nome_usuario_seguidor = seguidor.screen_name
        localizacao_seguidor = seguidor.location
        descricao_seguidor = seguidor.description
        num_seguidores_seguidor = seguidor.followers_count
        num_amigos_seguidor = seguidor.friends_count

        # escreva as informa√ß√µes do seguidor no arquivo CSV
        escritor_csv.writerow([id_seguidor, nome_seguidor, nome_usuario_seguidor, localizacao_seguidor, descricao_seguidor, num_seguidores_seguidor, num_amigos_seguidor])
```

### **04. Arquivo JSON**

O **JSON** (*JavaScript Object Notation*) √© um formato de arquivo para manter e trocar informa√ß√µes leg√≠veis pelas pessoas. O arquivo cont√©m apenas texto e usa a extens√£o *.json*.

O JSON √© um formato que armazena informa√ß√µes estruturadas e √© principalmente usado para transferir dados entre um servidor e um cliente.

O arquivo √© basicamente uma alternativa simples e mais leve ao XML (*Extensive Markup Language*), que tem fun√ß√µes similares.

Para ler um documento JSON dispon√≠vel na web em Python, voc√™ pode usar a biblioteca `requests` para fazer uma requisi√ß√£o HTTP ao URL do documento JSON e, em seguida, a biblioteca `json` para analisar o conte√∫do da resposta JSON. Aqui est√° um exemplo de c√≥digo:

```python
import requests
import json

# URL do arquivo JSON
url = 'https://quotes.rest/qod.json'

# fazer uma requisi√ß√£o HTTP ao URL e obter a resposta
response = requests.get(url)

# verificar se a requisi√ß√£o foi bem-sucedida
if response.status_code == 200:
    # analisar o conte√∫do da resposta JSON
    data = json.loads(response.content)
    
    # imprimir o conte√∫do do arquivo
    print(data)
else:
    # imprimir uma mensagem de erro se a requisi√ß√£o falhou
    print(f'A requisi√ß√£o falhou com c√≥digo de status {response.status_code}.')
```

Para fazer isso, podemos usar o seguinte c√≥digo em Python:

```python
import requests
from bs4 import BeautifulSoup

# faz a requisi√ß√£o √† p√°gina da Wikipedia
response = requests.get('https://pt.wikipedia.org/wiki/Am%C3%A9rica_do_Sul')

# cria um objeto BeautifulSoup para parsear o HTML
soup = BeautifulSoup(response.content, 'html.parser')

# encontra a tabela com a lista de pa√≠ses
tabela_paises = soup.find('table', {'class': 'wikitable sortable'})

# encontra as linhas da tabela
linhas = tabela_paises.find_all('tr')

# percorre as linhas, extraindo o nome do pa√≠s de cada c√©lula da primeira coluna
paises = []
for linha in linhas[1:]:
    celulas = linha.find_all('td')
    nome_pais = celulas[0].text.strip()
    paises.append(nome_pais)

# imprime a lista de pa√≠ses
print(paises)
```

#### **Web Scraping**

**Web scraping** √© a pr√°tica de coletar automaticamente dados de uma p√°gina da web, analisando seu conte√∫do HTML. Essa t√©cnica √© usada para extrair informa√ß√µes espec√≠ficas de uma p√°gina web, como pre√ßos de produtos em lojas online, informa√ß√µes de contato de empresas ou dados de pre√ßos de a√ß√µes em sites financeiros.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.imdb.com/chart/top'

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

movies = soup.select('td.titleColumn')
ratings = soup.select('td.ratingColumn.imdbRating')

for i in range(len(movies)):
    movie_title = movies[i].find('a').text
    movie_year = movies[i].find('span', class_='secondaryInfo').text
    movie_rating = ratings[i].find('strong').text
    print(f'{i+1}. {movie_title} ({movie_year}) - Rating: {movie_rating}')
```

Neste exemplo, fazemos uma solicita√ß√£o HTTP para a p√°gina do **IMDb** que lista os 250 melhores filmes de todos os tempos. Usamos a biblioteca `BeautifulSoup` para analisar o conte√∫do HTML da p√°gina e encontrar os elementos HTML que cont√™m as informa√ß√µes que desejamos extrair. Em seguida, usamos o m√©todo `select()` do `BeautifulSoup` para selecionar os elementos que cont√™m os t√≠tulos dos filmes e as classifica√ß√µes.

Usamos um la√ßo para iterar sobre as listas de t√≠tulos e classifica√ß√µes e, para cada filme, usamos o m√©todo `find()` do `BeautifulSoup` para encontrar os elementos HTML que cont√™m o t√≠tulo, o ano e a classifica√ß√£o. Em seguida, usamos o m√©todo text para extrair o texto desses elementos e imprimimos as informa√ß√µes relevantes na tela.

O exemplo a seguir extrai informa√ß√µes de filmes do IMDb e armazena em um DataFrame:

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Faz uma solicita√ß√£o GET para a p√°gina da web
url = 'https://www.imdb.com/chart/top'
response = requests.get(url)

# Verifica se a solicita√ß√£o foi bem-sucedida
if response.status_code == 200:
    # Analisa o HTML com a biblioteca BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Encontra as se√ß√µes com os dados que queremos extrair
    table = soup.find('table', {'class': 'chart full-width'})
    rows = table.find_all('tr')

    # Extrai os dados da tabela e armazena em listas
    titles = []
    years = []
    ratings = []
    for row in rows[1:]:
        cols = row.find_all('td')
        title = cols[1].find('a').text
        year = cols[1].find('span', {'class': 'secondaryInfo'}).text
        rating = cols[2].text.strip()
        titles.append(title)
        years.append(year)
        ratings.append(rating)

    # Cria o DataFrame a partir das listas
    df = pd.DataFrame({'Title': titles, 'Year': years, 'Rating': ratings})

    # Imprime o DataFrame para verifica√ß√£o
    print(df)
else:
    print('N√£o foi poss√≠vel obter a p√°gina da web')
```

Aqui est√° outro exemplo de web scraping em Python que usa a biblioteca `requests` e a biblioteca `BeautifulSoup` para extrair informa√ß√µes sobre not√≠cias de tecnologia do site TechCrunch:

```python
import requests
from bs4 import BeautifulSoup

url = 'https://techcrunch.com/'

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

articles = soup.select('.post-block__title__link')

for article in articles:
    title = article.text.strip()
    link = article['href']
    print(f'{title} - {link}')
```

Neste exemplo, fazemos uma solicita√ß√£o HTTP para a p√°gina inicial do TechCrunch e usamos a biblioteca `BeautifulSoup` para analisar o conte√∫do HTML da p√°gina e encontrar os elementos HTML que cont√™m as informa√ß√µes de t√≠tulo e link de cada artigo.

Usamos o m√©todo `select()` do `BeautifulSoup` para selecionar os elementos HTML que cont√™m as informa√ß√µes relevantes e, em seguida, usamos o m√©todo `text` e `['href']` para extrair as informa√ß√µes de t√≠tulo e link, respectivamente.

Por fim, imprimimos o t√≠tulo e link de cada artigo na tela.

```python
import requests
from bs4 import BeautifulSoup
import re

# URL do livro no Projeto Gutenberg
# Dom Casmurro , de Machado de Assis
url = 'https://www.gutenberg.org/files/55752/55752-h/55752-h.htm'

# Faz a solicita√ß√£o HTTP para obter o conte√∫do HTML da p√°gina do livro
response = requests.get(url)

# Cria um objeto BeautifulSoup para analisar o conte√∫do HTML
soup = BeautifulSoup(response.content, 'html.parser')

# Encontra o elemento HTML que cont√©m o texto do livro
book_text = soup.find('body')

# Extrai o texto do livro a partir do elemento HTML
book = book_text.get_text()

# Conta o n√∫mero de palavras que terminam com "mento"
words_amento = re.findall(r'\b\w+mento\b', book)
count_amento = len(words_amento)

print(f'O n√∫mero de palavras que terminam com "mento" √© {count_amento}.')
```

```python
import requests
import re

# URL do livro no Projeto Gutenberg
url = 'https://www.gutenberg.org/cache/epub/55752/pg55752.txt'

# Faz o download do livro em formato TXT
response = requests.get(url)
book = response.text

# Busca por todas as sequ√™ncias de quatro d√≠gitos no texto do livro
all_years = re.findall(r'\b\d{4}\b', book)

# Filtra apenas os anos v√°lidos (entre 1500 e 2023, por exemplo)
valid_years = [y for y in all_years if 1500 <= int(y) <= 2023]

# Remove anos duplicados
unique_years = set(valid_years)

# Ordena os anos em ordem crescente
sorted_years = sorted(unique_years)

print(f'Todos os anos que aparecem no livro s√£o: {sorted_years}.')
```

### **06. Arquivo CSV**

**CSV** (*Comma-Separated Values*) √© um formato de arquivo para armazenar dados tabulares, em que cada linha do arquivo representa uma linha da tabela e os valores s√£o separados por v√≠rgulas (ou outro caractere delimitador).

Em um arquivo CSV, cada linha representa um registro da tabela e cada coluna representa um campo. As informa√ß√µes s√£o organizadas em um formato de texto simples, com valores separados por v√≠rgulas (ou ponto-e-v√≠rgula, espa√ßo em branco ou outro caractere) para separar as colunas.

O formato CSV √© comum para compartilhar dados entre diferentes aplicativos e sistemas, porque √© simples e f√°cil de ler e escrever. Ele √© amplamente usado em planilhas, bancos de dados, ferramentas de an√°lise de dados e outras aplica√ß√µes que lidam com dados tabulares. Os arquivos CSV podem ser abertos e editados em muitos programas, incluindo Microsoft Excel, Google Sheets, LibreOffice Calc e muitos outros.

```python
import csv

# abre o arquivo CSV
with open('/content/example.csv', newline='') as arquivo_csv:
    # cria um objeto de leitor CSV
    leitor_csv = csv.reader(arquivo_csv, delimiter=',')
    # itera sobre cada linha no arquivo CSV
    for linha in leitor_csv:
        # exibe a linha na tela
        print(linha)
```

Neste exemplo, o c√≥digo usa a biblioteca `csv` do Python para abrir e ler um arquivo chamado "exemple.csv". Em seguida, ele itera sobre cada linha do arquivo usando um objeto de leitor CSV e exibe a linha na tela.

Veja um exemplo um pouco mais complexo de como ler um arquivo CSV e armazenar seus dados em uma lista de dicion√°rios em Python:

```python
import csv

# abre o arquivo CSV
with open('/content/example.csv', newline='') as arquivo_csv:
    # cria um objeto de leitor CSV
    leitor_csv = csv.DictReader(arquivo_csv)
    # inicializa uma lista vazia para armazenar os dados
    dados = []
    # itera sobre cada linha no arquivo CSV
    for linha in leitor_csv:
        # adiciona a linha como um dicion√°rio √† lista de dados
        dados.append(linha)

# exibe a lista de dados na tela
print(dados)
```

Neste exemplo, o c√≥digo usa `csv.DictReader` para ler o arquivo CSV e converter cada linha em um dicion√°rio Python com chaves correspondentes aos cabe√ßalhos das colunas no arquivo CSV. Ele ent√£o adiciona cada dicion√°rio a uma lista chamada dados.

Depois que todos os dados s√£o lidos e armazenados, o c√≥digo exibe a lista de dados na tela. Observe que, neste exemplo, os dados s√£o armazenados como uma lista de dicion√°rios, o que pode ser √∫til se voc√™ precisar manipular os dados em Python de forma mais avan√ßada do que apenas imprimir na tela.

Aqui est√° outro exemplo em que vamos criar um arquivo CSV e escrever nele usando Python:

```python
import csv

# abre o arquivo CSV em modo de grava√ß√£o
with open('output.csv', 'w', newline='') as arquivo_csv:
    # cria um objeto de escritor CSV
    escritor_csv = csv.writer(arquivo_csv)
    # escreve os cabe√ßalhos das colunas
    escritor_csv.writerow(['Nome', 'Idade', 'Cidade'])
    # escreve algumas linhas de dados
    escritor_csv.writerow(['Jo√£o', 25, 'S√£o Paulo'])
    escritor_csv.writerow(['Maria', 30, 'Rio de Janeiro'])
    escritor_csv.writerow(['Pedro', 20, 'Belo Horizonte'])
    escritor_csv.writerow(['Laura', 45, 'S√£o Paulo'])
    escritor_csv.writerow(['Sofia', 18, 'Rio de Janeiro'])
    escritor_csv.writerow(['Paulo', 42, 'Bahia'])

# exibe uma mensagem confirmando que o arquivo foi criado e gravado
print('Arquivo CSV criado e gravado com sucesso.')
```

Neste exemplo, o c√≥digo usa `csv.writer` para criar um objeto de escritor CSV e escrever dados em um arquivo chamado "output.csv". Primeiro, ele escreve uma linha com os cabe√ßalhos das colunas e, em seguida, escreve tr√™s linhas com dados.

Observe que, quando abrimos o arquivo CSV em modo de grava√ß√£o `('w')`, qualquer conte√∫do existente no arquivo ser√° exclu√≠do. Se voc√™ quiser adicionar novos dados a um arquivo existente, pode abrir o arquivo em modo de adi√ß√£o `('a')` em vez disso.

Aqui est√° outro exemplo em que usamos o m√≥dulo csv do Python para escrever dados em um arquivo CSV a partir de uma lista de dicion√°rios:

```python
import csv

# dados que ser√£o gravados no arquivo CSV
dados = [
    {'Nome': 'Jo√£o', 'Idade': 25, 'Cidade': 'S√£o Paulo'},
    {'Nome': 'Maria', 'Idade': 30, 'Cidade': 'Rio de Janeiro'},
    {'Nome': 'Pedro', 'Idade': 20, 'Cidade': 'Belo Horizonte'}
]

# abre o arquivo CSV em modo de grava√ß√£o
with open('output2.csv', 'w', newline='') as arquivo_csv:
    # cria um objeto de escritor CSV com cabe√ßalhos baseados nas chaves do primeiro dicion√°rio
    campos = dados[0].keys()
    escritor_csv = csv.DictWriter(arquivo_csv, fieldnames=campos)
    
    # escreve os cabe√ßalhos das colunas
    escritor_csv.writeheader()
    
    # escreve as linhas de dados
    for linha in dados:
        escritor_csv.writerow(linha)

# exibe uma mensagem confirmando que o arquivo foi criado e gravado
print('Arquivo CSV criado e gravado com sucesso.')
```

### **08. Trabalhos de Conclus√£o de Curso da UFSC**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Criando o dataframe
COLUNAS = [
   
    'Year',
    'Title',
    'Author',
    'Advisor',
    'Date',
    'Abstract',
    'Language',
    'Keywords',
    'PDF'
]

df = pd.DataFrame(columns=COLUNAS)

offset = 0
n = 1

# Curso de Antropologia
TCC_CURSO = 'https://repositorio.ufsc.br/handle/123456789/7487'
TCC_TOTAL = 35 

while offset <= TCC_TOTAL:
  
    url = TCC_CURSO + '/recent-submissions?offset=' + str(offset)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    links = soup.select('.ds-artifact-list a')

    for link in links:

        print(n)

        url_tcc = "https://repositorio.ufsc.br" + link.get('href')
        page = requests.get(url_tcc)
        soup = BeautifulSoup(page.content, 'html.parser')

        # Extraindo as informa√ß√µes
        title = soup.find('meta', attrs={'name': 'DC.title'})
        if title == None: 
          title = '' 
        else: 
          title = title['content']

        author = soup.find('meta', attrs={'name': 'DC.creator'})
        if author == None: 
          author = '' 
        else: 
          author = author['content']

        advisor = soup.find('meta', attrs={'name': 'DC.contributor'}).findNext('meta', attrs={'name': 'DC.contributor'})
        if advisor == None: 
          advisor = '' 
        else: 
          advisor = advisor['content']

        date = soup.find('meta', attrs={'name': 'DCTERMS.available'})
        if date == None: 
          date = '' 
        else: 
          date = date['content']

        year = date[0:4]

        abstract = soup.find('meta', attrs={'name': 'DCTERMS.abstract'})
        if abstract == None: 
          abstract = '' 
        else: 
          abstract = abstract['content']

        language = soup.find('meta', attrs={'name': 'DC.language'})
        if language == None: 
          language = '' 
        else: 
          language = language['content']

        keywords = soup.find('meta', attrs={'name': 'citation_keywords'})
        if keywords == None: 
          keywords = '' 
        else: 
          keywords = keywords['content']

        pdf = soup.find('meta', attrs={'name': 'citation_pdf_url'})
        if pdf == None: 
          pdf = '' 
        else: 
          pdf = pdf['content']

        new_row = {'Year': year, 'Title': title, 'Author': author, 'Advisor': advisor, 'Date': date, 
                   'Abstract': abstract, 'Language': language, 'Keywords': keywords, 'PDF': pdf}
        #print(new_row)
        
        df = df.append(new_row, ignore_index=True)

        n += 1
        
    offset += 20

# Exportando para Excel
df.to_excel('ufsc_tcc_Antropologia.xlsx')
```

Ele tamb√©m suporta recursos como hiperlinks, marcadores, notas e √≠ndices, tornando-o ideal para publica√ß√µes eletr√¥nicas como livros, revistas, jornais e outros tipos de conte√∫do.

Aqui est√° um exemplo b√°sico de como criar um arquivo EPUB em Python usando a biblioteca `epublib`:

```python
from ebooklib import epub

# Criar um objeto de livro EPUB
book = epub.EpubBook()

# Definir o t√≠tulo do livro
book.set_title('Meu Livro EPUB')

# Definir o nome do autor
book.add_author('Alexandre Donizeti Alves')

# Criar um cap√≠tulo e adicionar o conte√∫do
c1 = epub.EpubHtml(title='Cap√≠tulo 1', file_name='capitulo1.xhtml', lang='pt')
c1.content=u'<h1>Cap√≠tulo 1</h1><p>Este √© o conte√∫do do cap√≠tulo 1</p>'

# Adicionar o cap√≠tulo ao livro
book.add_item(c1)

# Criar uma tabela de conte√∫dos
book.toc = (epub.Link('capitulo1.xhtml', 'Cap√≠tulo 1', 'cap1'),)

# Adicionar a tabela de conte√∫dos ao livro
book.add_item(epub.EpubNcx())
book.add_item(epub.EpubNav())

# Gerar o arquivo EPUB
epub.write_epub('meu_livro.epub', book, {})
```

Neste exemplo, criamos um objeto `EpubBook` para representar nosso livro EPUB. Em seguida, definimos o t√≠tulo e o autor do livro. Em seguida, criamos um objeto `EpubHtml` para representar o conte√∫do do cap√≠tulo 1 e adicionamos esse cap√≠tulo ao livro. Finalmente, criamos uma tabela de conte√∫dos e adicionamos a NCX (√≠ndice hier√°rquico) e a navega√ß√£o do livro. Em seguida, usamos a fun√ß√£o `write_epub()` para gerar o arquivo EPUB e salv√°-lo no disco.

Por fim, imprimimos o t√≠tulo do livro usando a fun√ß√£o `print()`.

Aqui est√° um exemplo de c√≥digo em Python usando a fun√ß√£o `get_items_of_type()` da biblioteca `ebooklib` para extrair o conte√∫do de todos os cap√≠tulos de um arquivo no formato EPUB:

```python
import ebooklib
from ebooklib import epub

# Abrir o arquivo EPUB
book = epub.read_epub('meu_livro.epub')

# Encontrar todos os itens do tipo documento (cap√≠tulos)
chapter_items = book.get_items_of_type(ebooklib.ITEM_DOCUMENT)

# Percorrer cada item do cap√≠tulo e extrair o conte√∫do
for chapter_item in chapter_items:
    # Obter o conte√∫do do cap√≠tulo
    chapter_content = chapter_item.get_content()

    # Fazer algo com o conte√∫do (por exemplo, imprimir na tela)
    print(chapter_content.decode('utf-8'))
```

Aqui est√° um exemplo de c√≥digo em Python que utiliza a mesma abordagem anterior, mas que extrai o texto de cada cap√≠tulo sem utilizar as tags HTML do EPUB:

```python
from ebooklib import epub
from bs4 import BeautifulSoup

# Abrir o arquivo EPUB
book = epub.read_epub('meu_livro.epub')

# Encontrar todos os itens do tipo documento (cap√≠tulos)
chapter_items = book.get_items_of_type(ebooklib.ITEM_DOCUMENT)

# Percorrer cada item do cap√≠tulo e extrair o conte√∫do em formato de texto
for chapter_item in chapter_items:
    # Obter o conte√∫do do cap√≠tulo
    chapter_content = chapter_item.get_content()

    # Converter o conte√∫do em HTML para um objeto BeautifulSoup
    soup = BeautifulSoup(chapter_content, 'html.parser')

    # Extrair o texto do cap√≠tulo
    chapter_text = soup.get_text()

    # Fazer algo com o texto (por exemplo, imprimir na tela)
    print(chapter_text)
```

Neste exemplo, utilizamos a biblioteca `BeautifulSoup` para extrair o texto de cada cap√≠tulo do EPUB, em vez de utilizar as tags HTML diretamente. Primeiro, convertemos o conte√∫do do cap√≠tulo em HTML para um objeto `BeautifulSoup` usando o m√©todo `BeautifulSoup()` da biblioteca. Em seguida, utilizamos o m√©todo `get_text()` para extrair o texto sem as tags HTML. Por fim, fazemos algo com o texto (imprimimos na tela, por exemplo), assim como fizemos no exemplo anterior.

Para mostrar o conte√∫do do livro sem usar HTML e remover espa√ßos em branco desnecess√°rios, podemos utilizar a biblioteca `BeautifulSoup` para extrair somente o texto do conte√∫do e, em seguida, limpar os espa√ßos em branco com o m√©todo `strip()`.

```python
from ebooklib import epub
from bs4 import BeautifulSoup

# Abre o arquivo EPUB
book = epub.read_epub('meu_livro.epub')

# Obt√©m o conte√∫do do primeiro cap√≠tulo
chapter_content = book.get_item_with_href(book.toc[0].href).get_content()

# Converte o conte√∫do HTML em um objeto BeautifulSoup
soup = BeautifulSoup(chapter_content, 'html.parser')

# Obt√©m somente o texto do conte√∫do, sem tags HTML
text_content = soup.get_text()

# Remove espa√ßos em branco desnecess√°rios
text_content = ' '.join(text_content.split())

# Exibe o conte√∫do do cap√≠tulo sem tags HTML e sem espa√ßos em branco desnecess√°rios
print(text_content)
```

```python
import requests

# Define a URL do arquivo texto hospedado no Google Drive
url = 'https://drive.google.com/uc?id=1syswsF-6SfRrz3MTpZplu7Q9cisHBj6d&export=download'

# Faz uma solicita√ß√£o HTTP GET para obter o conte√∫do do arquivo
response = requests.get(url)

# Converte o conte√∫do da resposta em uma string do Python
text = response.content.decode('utf-8')

# Imprime o conte√∫do do arquivo na tela
print(text)
```
